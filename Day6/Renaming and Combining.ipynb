# Introduction
Oftentimes data will come to us with column names, index names, </br>
or other naming conventions that we are not satisfied with. In that case, </br>
you'll learn how to use pandas functions to change the names of the offending entries to something better.

We'll also explore how to combine data from multiple DataFrames and/or Series.

## RenamingÂ¶
The first function we'll introduce here is `rename()`, </br>
which lets you change index names and/or column names. For example, </br>
to change the `Sex` column in our dataset to `Gender`, we would do:

```python
titanic_df.rename(columns = {'Sex': 'Gender'})
```

`rename()` lets you rename index or column values by specifying a index or column keyword parameter,</br>
respectively. It supports a variety of input formats, but usually a Python dictionary is the most convenient.</br>
Here is an example using it to rename some elements of the index.

```python
titanic_df.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

You'll probably rename columns very often,</br>
but rename index values very rarely. For that, `set_index()` is usually more convenient.</br>

Both the row index and the column index can have their own name attribute.</br>
The complimentary `rename_axis()` method may be used to change these names. For example:

```python
titanic_df.rename_axis("Row_index", axis='rows').rename_axis("Column_index", axis='columns')
```

## CombiningÂ¶

When performing operations on a dataset, </br>
we will sometimes need to combine different DataFrames and/or Series in non-trivial ways.</br>
Pandas has three core methods for doing this. </br>
In order of increasing complexity, these are `concat()`, `join()`, and `merge()`. </br>
Most of what `merge()` can do can also be done more simply with `join()`, </br>
so we will omit it and focus on the first two functions here.

The simplest combining method is `concat()`. Given a list of elements, </br>
this function will smush those elements together along an axis.

This is useful when we have data in different DataFrame or Series objects but having the same fields (columns).</br>
`One example:` the YouTube Videos dataset,</br>
which splits the data up based on country of origin (e.g. Canada and the UK, in this example).</br>
If we want to study multiple countries simultaneously,</br>
we can use `concat()` to smush them together:

```python
canadian_youtude = pd.read_csv("trending youtude video/CAvideos.csv")
british_youtude = pd.read_csv("trending youtude video/GBvideos.csv")
pd.concat([canadian_youtude, british_youtude])
```

The middlemost combiner in terms of complexity is `join()`.</br>
`join()` lets you combine different DataFrame objects which have an index in common.</br>
For example, to pull down videos that happened to be trending on the same day in both Canada and the UK,</br>
we could do the following:

```python
left = canadian_youtude.set_index(['title', 'trending_date'])
right = british_youtude.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```

The `lsuffix` and `rsuffix` parameters are necessary here because the data has the same column names in both British and Canadian datasets. </br>
If this wasn't true (because, say, we'd renamed them beforehand) we wouldn't need them.
